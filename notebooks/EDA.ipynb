```python
import pandas as pd

train_df = pd.read_csv('data/train.csv')
train_df.head()
```

```python
train_df.groupby('category')['id'].count().sort_values(ascending=False)
```

```python
import numpy as np
import matplotlib.pyplot as plt


plt.rcParams["font.family"] = 'NanumGothic'

nums_category = train_df.groupby('category')['id'].count().sort_values(ascending=False)
x = nums_category.values
percent = x / x.sum() * 100.0
labels = nums_category.index

plt.figure(figsize=(8, 10))
plt.pie(nums_category.values, labels=labels, autopct=lambda pct: f'{pct:.2f}% - {int(np.round(pct / 100. * x.sum()))}')
plt.show()
```

```python
length_answers = train_df[['답변_1', '답변_2', '답변_3', '답변_4', '답변_5']].map(len).values.flatten()
```

```python
print('max:', length_answers.max())
print('min:', length_answers.min())
print('mean:', length_answers.mean())
print('std:', length_answers.std())
```

```python
plt.hist(length_answers)
plt.title('답변 길이 히스토그램')
plt.xlabel('길이')
plt.ylabel('빈도')
plt.show()
```

```python
length_questions = train_df[['질문_1', '질문_2']].map(len).values.flatten()

print('max:', length_questions.max())
print('min:', length_questions.min())
print('mean:', length_questions.mean())
print('std:', length_questions.std())
```

```python
plt.hist(length_questions)
plt.title('질문 길이 히스토그램')
plt.xlabel('길이')
plt.ylabel('빈도')
plt.show()
```
